What is NumPy?

NumPy is a Python library used for working with arrays. It also has functions for working in domain of linear algebra, fourier transform, and matrices. NumPy stands for Numerical Python.

Why Use NumPy?
In Python we have lists that serve the purpose of arrays, but they are slow to process.

NumPy aims to provide an array object that is up to 50x faster than traditional Python lists.

The array object in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy.

Arrays are very frequently used in data science, where speed and resources are very important.

```python
"""
create a definition
"""

def get_stats(x):

    """
    args: <array>
    outputs: min, max, mean
    
    """
    min_, max_, mean_ = np.min(x), np.max(x), np.mean(x)
    return min_, max_, mean_
    
```

```python
""" example """
x = np.array([0,1,2,3,4,5])
min_, max_, mean_ = get_stats(x)
print("Minimum:",min_,"\nMaximum:",max_,"\nMean:",mean_)
```

```python
""" With the help of Scipy and Matplotlib.pyplot we will:

1. Generate a gaussian distribution of random numbers using Numpy 
2. Fit that distribution using Scipy https://www.w3schools.com/python/scipy/scipy_intro.php
3. Plot the results with pyplot

"""
```

```python
# Generate gaussian distribution

from numpy import random
g_dist = random.normal(size=(10000,),
              loc = 0,
              scale=1)
counts, bins = np.histogram(g_dist,bins=20)
```

```python
bins = (bins[:-1] + bins[1:]) / 2
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,5))
ax.bar(bins,counts/np.sum(counts),
         color = 'black',
         alpha = 0.5,
         width = 0.35,
         label = 'data')

ax.set_title('Gaussian',fontsize=15)
ax.set_ylabel('Normalized Counts')
```

$$ f(x) = \frac{1}{\sigma\sqrt{2\pi}} \exp^{-(x-\mu)^2/2\sigma^2}$$
$$\sigma = \text{Standard Deviation}  $$
$$\mu = \text{Mean}$$

```python
""" 
Fit the distribution 
"""

# create the function

gaussian_func = lambda x: x[2] *np.exp( - ( bins - x[1] )**2 / (2*x[0]**2))

""" 
lambda x: this syntax is a common way to create functions of numeric variables, it has many use cases we will see this again
x[0] = standard deviation
x[1] = mean
x[2] = amplitude
'**2' = ^2
"""
```

$$ R(x) = \left( f(x) - \hat{y}\right)^2$$

```python
# create a function to minimize

resid = lambda x: np.dot((gaussian_func(x) - counts),(gaussian_func(x) - counts))
```

$$ \begin{align}
f(x) &= f(x_0) + (x-x_0)\frac{df}{dx}|_{x_0}+...\\
f(x)&\approx  f(x_0) + (x-x_0)\frac{df}{dx}|_{x_0} \\
\Rightarrow x_{n+1} &= x_n - \frac{f(x_n)}{\frac{df}{dx}|_{x_n}}
\end{align}$$

```python
""" minimize the residuals """

from scipy.optimize import minimize

initial_guess = [1,3,1e3]
result = minimize(resid, initial_guess,tol=1e-3)

print("Parameter results:",result.x)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(bins,counts,
         color = 'black',
         alpha = 0.5,
         width = 0.35,
         label = 'data')

ax.set_title('Gaussian',fontsize=15)
ax.set_ylabel('Counts',fontsize = 12)
ax.plot(bins,
        gaussian_func(result.x),
        color = 'blue',
        marker = 'o',
        label = 'fit')
ax.legend(loc = 'upper left', fontsize= 12)
```
